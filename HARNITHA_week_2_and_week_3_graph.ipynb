Brute Force Tag Recommendation Weighted Graph

LIBRARIES USED:
REQUESTS - used for web scraping and handle requests to webpapges

tagsList- store list of values from each tag 
res- requests object used to get data from website using GET 
jsonPage- store data in json format
title - store all the values of key items 

```

import requests


tagsList=[]
#for to extract all tags from 100 pages
for i in range(1,101):
  res=requests.get('https://api.stackexchange.com/2.2/questions?page='+str(i)+'&pagesize=100&order=desc&sort=activity&site=datascience')
  jsonPage=res.json()
  title=jsonPage['items']
  for tag in title:
    #store each set of tags as a nested list
    tagsList.append(tag['tags'])

```

Getting the unique tags from the total list of tags

Retrieve all the tag names and store it in a list.
use set method to find the unique tags

```
count=0
totalTags=[]
#collect all tags in all the tag sets 
#use set method to find unique tags
for tag in tagsList:
  for ele in tag:
    count+=1
    totalTags.append(ele)
uniqueTags=set(totalTags)
uniqueTagList=list(uniqueTags)
```

Forming a list of edges:


for each tag in a set of tags , is mapped to each a every tag in uniqueTagList

ele!=unique is used to avoid self loops
# New Section
.sort() is used to find repeated nodes easily

temp- a temporary list that combines start and end node
edges- list that contains all edges formed

```

#for each tag in a set of tags , is mapped to each a every tag in uniqueTagList

#ele!=unique is used to avoid self loops
#.sort() is used to find repeated nodes easily

#temp- a temporary list that combines start and end node
#edges- list that contains all edges formed

temp=[]
edges=[]
for row in tagsList:
  for i in range(0,len(row)-1):
    for j in range(i,len(row)):
      if row[i]!=row[j]:
        temp.append(row[i])
        temp.append(row[j])
        temp.sort()
        edges.append([temp[0],temp[1]])
        temp.clear()
len(edges)
    

```

Forming a graph using networkX

```
import networkx as nx
import matplotlib.pyplot as plt
import pickle as pk

# create a graph
G=nx.Graph()
# intialize a new egde with weight 1
#if edge already exist,increment weight
for edge in edges:
  if G.has_edge(edge[0],edge[1]):
    G[edge[0]][edge[1]]['weight']+=1
  else:
    G.add_edge(edge[0],edge[1],weight=1)



elarge=[(u,v) for (u,v,d) in G.edges(data=True) if d['weight'] >1]
esmall=[(u,v) for (u,v,d) in G.edges(data=True) if d['weight'] ==1]

pos=nx.spring_layout(G) # positions for all nodes

#pickling and serializing the graph- to used in flask app 
nx.write_gpickle(G,'graph.gpickle')
```

```
# take input from user
given_tag=input()
# find neighbors of the node(tag)
related=list(G.neighbors(given_tag))
neigh_weighted=[]
temp=[]

#get the edge weights for each neighbour node
for ele in related:
  weight=G[given_tag][ele]['weight'] 
  neigh_weighted.append([ele,weight])
#sort accoring to weights
sorted_neigh=sorted(neigh_weighted,key=lambda x:x[1])

#extract only the tags
list_of_related_tags=[]
for tag in sorted_neigh:
  list_of_related_tags.append(tag[0])
# display the tags
print(list_of_related_tags)
```
